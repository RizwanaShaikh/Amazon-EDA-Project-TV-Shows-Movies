## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Import necessary libraries
import pandas as pd
import numpy as np

# Load datasets
titles_file_path = "C:/Users/afridi mansoori/Desktop/Amazon EDA/Amazon EDA/titles.csv"
credits_file_path = "C:/Users/afridi mansoori/Desktop/Amazon EDA/Amazon EDA/credits.csv"

titles_df = pd.read_csv(titles_file_path)
credits_df = pd.read_csv(credits_file_path)

# Display basic info before cleaning
print("Before Cleaning:\n")
print(titles_df.info())
print(credits_df.info())

# Remove duplicates
titles_df = titles_df.drop_duplicates()
credits_df = credits_df.drop_duplicates()

# Handling missing values in titles dataset
titles_df.loc[:, "description"] = titles_df["description"].fillna("No description available")
titles_df.loc[:, "age_certification"] = titles_df["age_certification"].fillna("Unknown")
titles_df.loc[:, "seasons"] = titles_df["seasons"].fillna(0)  # Assuming missing seasons means it's a movie
titles_df.loc[:, "imdb_score"] = titles_df["imdb_score"].fillna(titles_df["imdb_score"].median())  # Median imputation
titles_df.loc[:, "tmdb_score"] = titles_df["tmdb_score"].fillna(titles_df["tmdb_score"].median())  # Median imputation
titles_df.loc[:, "imdb_votes"] = titles_df["imdb_votes"].fillna(0)  # Filling missing votes with 0
titles_df.loc[:, "tmdb_popularity"] = titles_df["tmdb_popularity"].fillna(0)  # Filling missing popularity with 0

# Handling missing values in credits dataset
credits_df.loc[:, "character"] = credits_df["character"].fillna("Unknown")

# Fill missing values in credits dataset
credits_df = credits_df.assign(character=credits_df["character"].fillna("Unknown"))

# Convert data types
titles_df["release_year"] = pd.to_numeric(titles_df["release_year"], errors="coerce")
titles_df["seasons"] = titles_df["seasons"].astype(int)  # Convert seasons to integer
titles_df["imdb_votes"] = titles_df["imdb_votes"].astype(int)  # Convert votes to integer

# Standardize string-based columns (removing extra spaces & converting to lowercase)
titles_df["title"] = titles_df["title"].str.strip().str.lower()
titles_df["genres"] = titles_df["genres"].str.strip().str.lower()
titles_df["production_countries"] = titles_df["production_countries"].str.strip().str.lower()

credits_df["name"] = credits_df["name"].str.strip().str.lower()
credits_df["role"] = credits_df["role"].str.strip().str.lower()
credits_df["character"] = credits_df["character"].str.strip().str.lower()

# Display basic info after cleaning
print("\nAfter Cleaning:\n")
print(titles_df.info())
print(credits_df.info())

# Save cleaned datasets for further analysis
titles_df.to_csv("C:/Users/afridi mansoori/Desktop/Amazon EDA/Amazon EDA/titles.csv", index=False)
credits_df.to_csv("C:/Users/afridi mansoori/Desktop/Amazon EDA/Amazon EDA/credits.csv", index=False)

print("\nData Cleaning Completed! Cleaned files saved successfully.")
````

````python
from ast import literal_eval
import pandas as pd

# Check if genres column is already a list or a string
def safe_literal_eval(val):
    try:
        return literal_eval(val) if isinstance(val, str) else val
    except (ValueError, SyntaxError):
        return []  # Return empty list if parsing fails

# Apply the function safely
titles_df["genres"] = titles_df["genres"].apply(safe_literal_eval)

# Flatten genre lists and count occurrences
all_genres = [genre for sublist in titles_df["genres"] for genre in sublist]
genre_counts = pd.Series(all_genres).value_counts()

# Convert to DataFrame for visualization
genre_df = pd.DataFrame({'Genre': genre_counts.index, 'Count': genre_counts.values})

# Plot genre distribution
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
ax = sns.barplot(data=genre_df.head(10), x="Genre", y="Count", hue="Genre", palette="viridis", legend=False)
plt.xticks(rotation=45, ha="right")
plt.title("Most Common Genres in Amazon Prime Content", fontsize=14)
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.bar_label(ax.containers[0])  # Add labels on bars
plt.show()
````
